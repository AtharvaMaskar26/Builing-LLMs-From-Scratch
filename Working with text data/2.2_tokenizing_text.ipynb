```python
import re

text = "Hello, world. Is this -- a test?"

# Corrected regex
result = re.split(r'([,.:;?_!"()\s-]+)', text)
result = [item.strip() for item in result if item.strip()]

print(result)
```

```python
# Applying this to the raw text

preprocessed =  re.split(r'([,.:;?_!"()\s-]+)', raw_text)
preprocessed = [item.strip() for item in preprocessed if item.strip()]

print(len(preprocessed))
print(preprocessed[:30])
```

```python
# Creating a vocabulary 

all_words = sorted(list(set(preprocessed)))

vocab_size = len(all_words)
print(vocab_size)
```

```python
vocabulary = {index:vocab for index, vocab in enumerate(all_words)}

for i, item in vocabulary.items():
    print(i, item)

    if i > 50:
        break
```

```python
class SimpleTokenizerV1:
    def __init__(self, vocab):
        self.str_to_int =  vocab
        self.int_to_str = {v:i for i, v in vocab.items()}

    def encode(self, text: str):
        preprocessed =  re.split(r'([,.:;?_!"()\s-]+)', text)
        preprocessed = [item.strip()  for item in preprocessed if item.strip()]

        ids = [self.str_to_int[s] for s in preprocessed]

        return ids

    def decode(self, ids):
        text = " ".join(self.int_to_str[i] for i in ids)

        text = re.sub(r'([,.:;?_!"()\s-]+)', text)

        return text
    
tokenizer = SimpleTokenizerV1(vocab=vocabulary)

text = """It's the last he painted, you know, " Mrs. Gisburn said"""

ids = tokenizer.encode(text)

print(ids)

decoded_text = tokenizer.decode(ids)

print(decoded_text)
```

```python
class SimpleTokenizerV2:

    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {v:i for i, v in vocab.items()}
    
    def encode(self, text):
        preprocessed =  re.split(r'([,.:;?_!"()\s-]+)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [item if item in self.str_to_int else "<|unk|>" for item in preprocessed]

        ids = [self.str_to_int[s] for s in preprocessed]

        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])

        text = re.sub(r'([,.:;?_!"()\s-]+)', text)

        return text
    

text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the place"

text = "<|endoftext|>".join((text1, text2))

print(text)
```
